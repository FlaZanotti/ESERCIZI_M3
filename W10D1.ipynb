## ESERCIZIO 6

Scrivere un programma che, data una lista di numeri, fornisca in output i tre numeri più grandi;  
gestire il caso in cui la lista sia più corta di tre, e quando uno o più dei numeri selezionati sono uguali.

```python
numeri = [5, 10, 60, 40, 3, 71, 45, 53, 21, 80, 80, 96, 14, 12]
numeriset = set(numeri) #trasformo in set così elimino i duplicati
numeri = list(numeriset) #trasformo in lista così è ordinabile con sort()
if len(numeri) > 3:
    numeri.sort()
    print(numeri[-3:])
else:
    print("La lista è più corta di 3:")
    print(numeri)
```

## ESERCIZIO 7

Scrivere un programma che:
- in input acquisisce una lista di numeri e un numero K
- in output, dovrà restituire la media di tutti i numeri nella lista maggiori o uguali a K
- se non ce ne dovesse essere nessuno, dovrà stampare a schermo un messaggio adeguato.

```python
import pandas as pd
```

```python
lista_numeri = [5, 10, 60, 40, 3, 71, 45, 53, 21, 80, 80, 96, 14, 12]
K = int(input("Inserire numero:"))
lista_numeri = pd.Series(lista_numeri)
lista_numeri_big = lista_numeri[lista_numeri >= K]

if not lista_numeri_big.empty:
    print (lista_numeri_big.mean())
else:
    print("Nessun valore soddisfa la condizione")
```

## ESERCIZIO 12

Abbiamo una lista di studenti:

studenti = ["Alex", "Bob", "Cindy", "Dan", "Emma", "Faith", "Grace", "Henry", "Isabelle", "John"]  


vogliamo dividere gli studenti in due squadre per un campionato di Uno nel seguente modo:
- selezioneremo i nomi in posizione pari per un squadra
- i nomi in posizione dispari per l'altra.
Creiamo due liste per ogni squadra, e alla fine visualizziamole.

```python
studenti = ["Alex", "Bob", "Cindy", "Dan", "Emma", "Faith", "Grace", "Henry", "Isabelle", "John"]
```

```python
print (studenti[1])
```

```python
squadra_pari = []
squadra_dispari = []

for i, nome in enumerate(studenti): #conto le posizioni degli elementi nella lista
    if i % 2 == 0:
        squadra_pari.append(nome)
    else:
        squadra_dispari.append(nome)

print("Squadra pari:", squadra_pari)
print("Squadra dispari:", squadra_dispari)
```
